## Wider Dynamic Range of Single Precision

`tf.keras.backend.ones` instantiates an all-ones variable and returns it.

Below we use `tf.keras.backend.ones` to instantiate matrices of dimensions (64, 64) with all entries having the value 16.

Update the `dtype` below such that half precision is used for `half_var`.

```
# half_var = 16*tf.keras.backend.ones((64,64), dtype=None)
half_var = 16*tf.keras.backend.ones((64,64), dtype='float16')
```

Update the `dtype` below such that single precision is used for `single_var`

```
# single_var = 16*tf.keras.backend.ones((64,64), dtype=None)
single_var = 16*tf.keras.backend.ones((64,64), dtype='float32')
```

The entries can be summed up with `tf.keras.backend.sum` and we observe the results produced for the different precisions.

```
tf.keras.backend.sum(
    half_var, axis=None, keepdims=False)
```

```
tf.keras.backend.sum(
    single_var, axis=None, keepdims=False)
```

## Increased Precision of Single Precision

```
param = 1
update = 0.0001
```

```
half_param = tf.keras.backend.variable(value=param, dtype='float16')
half_update = tf.keras.backend.variable(value=update, dtype='float16')
```

```
half_param + half_update
```

```
single_param = tf.keras.backend.variable(value=param, dtype='float32')
single_update = tf.keras.backend.variable(value=update, dtype='float32')
```

```
single_param + single_update
```
